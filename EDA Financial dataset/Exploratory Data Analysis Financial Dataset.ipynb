# Exploratory Data Analysis of a Financial Dataset

## Introduction

The primary objective of this exploratory data analysis is to gain insights into the dataset's financial metrics, understand the distribution and relationships among the variables, and particularly investigate the factors influencing the Price to Earnings (PE) ratio.

## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Here is a glossary description for each of the 37 columns in the dataset:

•	**permno**: A unique permanent identifier assigned to each security by CRSP (Center for Research in Security Prices).

•	**adate**: The date of the annual financial report (refers to the fiscal year end).

•	**qdate**: The date of the quarterly financial report.

•	**public_date**: The date when the financial data was made public.

•	**pe_inc**: Price to earnings ratio, calculated using the net income.

•	**npm**: Net profit margin; it measures how much of each dollar in revenue is translated into profit.

•	**opmbd**: Operating margin before depreciation, indicating profitability from operations before the impact of depreciation.

•	**opmad**: Operating margin after depreciation, indicating profitability from operations after accounting for depreciation.

•	**gpm**: Gross profit margin; the difference between revenue and cost of goods sold divided by revenue.

•	**ptpm**: Pretax profit margin, a measure of a company's earnings before tax as a percentage of total sales or revenue.

•	**cfm**: Cash flow margin, an indicator of a company’s efficiency at converting its cash inflows into profit.

•	**roa**: Return on assets, a measure of how effectively the company uses its assets to generate earnings.

•	**roe**: Return on equity, indicating how effectively financial resources are being utilized to generate earnings.

•	**efftax**: Effective tax rate, the average rate at which a corporation's pre-tax profits are converted to taxes.

•	**aftret_invcapx**: After-tax return on invested capital excluding excess cash.

•	**aftret_equity**: After-tax return on equity, measuring profitability after taxes are accounted for.

•	**GProf**: Gross profitability, a measure of gross profits compared to total assets.

•	**debt_invcap**: Debt to invested capital ratio.

•	**totdebt_invcap**: Total debt to invested capital ratio.

•	**debt_at**: Debt to total assets ratio.

•	**curr_debt**: Current portion of long-term debt.

•	**lt_debt**: Long-term debt, excluding the current portion.

•	**cash_debt**: Cash to debt ratio, an indicator of a company's ability to cover its total debt with its cash balance.

•	**dltt_be**: Long-term debt to book equity ratio.

•	**debt_assets**: Total debt divided by total assets.

•	**debt_capital**: Total debt to total capital ratio.

•	**de_ratio**: Debt to equity ratio, indicating the proportion of equity and debt used to finance a company's assets.

•	**intcov**: Interest coverage ratio, a measure of a company's ability to meet its interest payments.

•	**intcov_ratio**: Another variant of interest coverage ratio.

•	**cash_ratio**: Cash ratio, a liquidity indicator showing a company's ability to cover its short-term obligations with its cash and cash equivalents alone.

•	**quick_ratio**: Quick ratio, a measure of a company's ability to meet its short-term obligations with its most liquid assets.

•	**curr_ratio**: Current ratio, a liquidity ratio that measures a company's ability to pay off its short-term liabilities with its short-term assets.

•	**inv_turn**: Inventory turnover, a ratio showing how many times a company’s inventory is sold and replaced over a period.

•	**at_turn**: Asset turnover, a measure of how efficiently a company uses its assets to generate sales or revenue.

•	**rect_turn**: Receivables turnover ratio, an indicator of how effectively a company uses and manages the credit it extends to customers and how quickly that short-term debt is collected or is paid.

•	**pay_turn**: Payables turnover ratio, a measure of how quickly a company pays off its suppliers.

•	**TICKER**: The stock ticker symbol for the company.


# Dataset Inspection via Heatmap

```python
# Filter numerical columns for correlation analysis
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numerical_data.corr()

# Generate a heatmap
plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', linewidths=0.1, linecolor='white')
plt.title('Correlation Heatmap')
plt.show()

# Distribution of a few selected metrics
fig, ax = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(data['pe_inc'], bins=50, ax=ax[0], kde=True, color='blue')
ax[0].set_title('Distribution of PE Ratio (pe_inc)')
sns.histplot(data['roe'], bins=50, ax=ax[1], kde=True, color='green')
ax[1].set_title('Distribution of Return on Equity (roe)')
sns.histplot(data['de_ratio'], bins=50, ax=ax[2], kde=True, color='red')
ax[2].set_title('Distribution of Debt to Equity Ratio (de_ratio)')
plt.tight_layout()
plt.show()
```

# Dataset Inspection via Box Plot

```python
# Creating box plots for each numerical feature
num_cols = numerical_data.columns
n_cols = 5
n_rows = (len(num_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 20))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(x=numerical_data[col], ax=axes[i], color='skyblue')
    axes[i].set_title(col, fontsize=10)
    axes[i].tick_params(axis='x', labelsize=8)

# Remove unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

fig.tight_layout(pad=2.0)
plt.show()
```

•	**Outliers**: Many features display a significant number of outliers, which could significantly affect the mean and standard deviation of these distributions. In financial data, outliers might represent companies with unique circumstances, such as those in different industries, of different sizes, or experiencing unusual events (e.g., mergers, acquisitions, large market movements).

•	**Skewness**: Many distributions appear to be right-skewed (positive skew), where the median is lower than the mean. This skewness indicates that while the bulk of the data clusters towards the lower end, there are a few very high values pulling the mean upwards.

•	**Scale and Variance**: The scales of the variables differ widely, suggesting that normalization or standardization might be necessary if you plan to use this data for machine learning models or other analyses where scale impacts the results.

•	**Potential Data Errors**: The presence of extreme outliers could also suggest data errors or anomalies that might need verification or exclusion from analysis to prevent biased results.

•	**Sector-Specific Trends**: Outliers and variability might reflect sector-specific trends. Companies in certain industries (like technology or finance) might have characteristics significantly different from those in sectors like manufacturing or utilities.


# Dataset Inspection via Scatter Plot

```python
# Remove 'pe_inc' from the features list to prevent plotting it against itself
features_for_scatter = [col for col in num_cols if col != 'pe_inc']

# Determine layout for the plots
n_cols = 5
n_rows = (len(features_for_scatter) + n_cols - 1) // n_cols

# Create scatter plots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))
axes = axes.flatten()

for i, feature in enumerate(features_for_scatter):
    axes[i].scatter(data[feature], data['pe_inc'], alpha=0.5, s=10)
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('pe_inc')
    axes[i].set_title(f'Scatter plot of pe_inc vs {feature}')

# Remove unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.tight_layout()
plt.show()
```

•	**PE Ratio Extremes**: The pe_inc feature shows some extremely high and negative values which are quite distant from the bulk of the data points. This could be indicative of outliers, possibly due to errors or exceptional cases like companies with very high or low earnings relative to their share price.

•	**Clustering and Spread**: In most plots, the data points are clustered in a specific region, indicating a concentration of values within a range. For example, metrics like npm, opmbd, and opmad have clusters near the lower values with fewer companies stretching towards higher PE ratios.

•	**Correlations**: Some variables appear to have a potential correlation with pe_inc. For example, profitability ratios (npm, opmad, gpm, cfm) seem to have a cluster of points around the lower pe_inc values, which could suggest a relationship where higher profitability ratios might correspond to a lower PE ratio.

•	**Financial Health Metrics**: Ratios such as debt_at, de_ratio, cash_ratio, and quick_ratio which give an indication of a company's financial health, show varied dispersion patterns suggesting different levels of association with the PE ratio.

•	**Efficiency Metrics**: Turnover ratios like inv_turn, at_turn, rect_turn, and pay_turn also show clusters but no clear linear pattern, suggesting that the efficiency of asset and inventory management might not have a direct linear relationship with the market valuation as indicated by PE ratio.


# Scatter Plot with Feature transformation

```python
# Adjusting the scatter plot visuals as requested
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))
axes = axes.flatten()

for i, feature in enumerate(features_for_scatter):
    axes[i].scatter(data[feature], data['pe_inc']**2, alpha=0.5, color='red', s=10)  # s is the size of the dot
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('pe_inc^2')
    axes[i].set_title(f'Scatter plot of pe_inc^2 vs {feature}')

# Remove unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.tight_layout()
plt.show()
```

Here's what these plots suggest:

•	**Concentration of Lower Values**: The majority of the pe_inc^2 values are close to zero, which is evident from the dense cluster of points near the origin in each plot. This indicates that for most data points, the PE ratio is relatively low.

•	**Presence of Extreme Values**: There are some extreme values of pe_inc^2 that significantly deviate from the main cluster. This is especially noticeable in variables that can take large positive values, such as various ratios and turnover rates.

•	**Sparse High-Value Data**: The data points with high pe_inc^2 values are quite sparse, indicating that only a small number of companies have a very high squared PE ratio.

•	**Lack of Clear Linear Relationship**: There doesn't seem to be a clear linear trend in any of the plots. Instead, most of the variability in pe_inc^2 is seen in the vertical spread of data points, indicating a potential non-linear relationship or no relationship.

•	**Impact of Squaring**: Squaring the PE ratio has amplified the effect of outliers, as the variation in higher PE ratio values has become more pronounced. This transformation makes it easier to identify outliers but may also skew analysis if not handled properly.
